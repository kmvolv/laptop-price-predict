```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Scraper-Data-Clean.csv', index_col = 0)
df.drop(['id'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.OS.value_counts() 
```

```python
sns.stripplot(data = df, x = "Price", y = "Brand", hue = "Type")
```

```python
sns.violinplot(data = df, x = "CPUbrand", y = "Price", hue = "Type")
```

```python
sns.barplot(data = df, x = "GPUbrand", y = "Price", hue = "Type")
```

```python
sns.boxplot(data = df, x = 'Weight', y = 'Price', hue = "Type", palette = 'dark')
```

```python
sns.lmplot(data = df, x = 'PPI', y = 'Price', hue = 'Type')
```
